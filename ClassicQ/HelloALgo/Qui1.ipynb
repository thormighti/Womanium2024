```python
from classiq import * 

@qfunc
def main(y: Output[QNum], x : Output[QNum], z: Output[QNum]):
    #allocate 5 qubits in super position
    allocate(5,x)
    allocate(5,z)
    #in super position
    hadamard_transform(x)
    hadamard_transform(z)
    y |= 2*x + 4*z + 2
```

```python
quantum_prog = synthesize(create_model(main))
```

```python
show(quantum_prog)
```

```python
job = execute(quantum_prog)
results = job.result()[0].value.parsed_counts
```
